```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Summer.csv')
df.head(10)
```

### 1. In how many cities Summer Olympics is held so far?

```python
 print("Number of cities in which Olympic was held so far:", len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
sport_medal = df[['Sport', 'Medal']]
sport_gold = {}
for i in sport_medal[sport_medal['Medal'] == 'Gold']['Sport']:
    if i in sport_gold:
        sport_gold[i] += 1
    else:
        sport_gold[i] = 1
        
df_sport_gold = pd.DataFrame.from_dict(sport_gold, orient='index', columns=['Number of Gold Medals']).sort_values(by='Number of Gold Medals')
df_sport_gold[-5:].plot(kind='barh', y='Number of Gold Medals', ylabel='Sport', title='Top-5 Sports with Most Number of Gold Medals', figsize=(10, 6))
plt.show()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sport_medal_tally = []
for i in sport_medal['Sport'].unique():
    sport_medal_tally.append([i, len(sport_medal[sport_medal['Sport']==i])])

df_sport_medal_tally = pd.DataFrame(sport_medal_tally, columns=['Sport', 'Number of Medals']).sort_values(by='Number of Medals')
df_sport_medal_tally[-5:].plot(kind='barh',x='Sport', y='Number of Medals', title='Top-5 Sports with Most Number of Medals', figsize=(10, 6))
plt.show()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
country_medal_year = df[['Country', 'Medal', 'Year']]
# country_medal_year

ind_medal_year = country_medal_year[country_medal_year['Country'] == 'IND']
# ind_medal_year

ind_gold_year = ind_medal_year[ind_medal_year['Medal'] == 'Gold'].sort_values(by='Year')
# ind_gold_year

print('India Won the First Gold Medal in', ind_gold_year['Year'].iloc[0], 'in Summer Olympics')
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
sport_athlete = df[['Sport', 'Athlete']]
athletes_per_sport = []
for i in sport_athlete['Sport'].unique():
    athletes_per_sport.append([i, len(sport_athlete[sport_athlete['Sport'] == i]['Athlete'])])

df_athletes_per_sport = pd.DataFrame(athletes_per_sport, columns=['Sport', 'Number of Players']).sort_values(by='Number of Players')
df_athletes_per_sport[-5:].plot(kind='barh', x='Sport', y='Number of Players', title='Top-5 Sports with Most Number of Players', figsize=(10, 6))
plt.show()
```
